## SoloVPE & Beer-Lambert Gesetz

SoloVPE ist ein UV-Vis-NIR Spektrometer. Es ermöglicht die präzise Bestimmung von Konzentrationen und Extinktions Koeffizienten mittels mehreren Absorbtions Messungen über unterschiedliche Weglängen. Eine ausführlichere Einführung kann hier [Slope Spectroscopy](https://www.repligen.com/ctech-resources/pdf/The-Power-of-Slope-Spectroscopy-1.pdf) gefunden werden. Das Beer-Lambert Gesetz liefert die physikalische Grundlage für die Funktion das Spektrometers:

$$ A = \alpha * l * c $$
* A: gemessene Absorbtion 
* $\alpha$ : molarer Absorbtions-Koeffizient (Wellenlängen-Abhängig)
* l: Weglänge 
* c: Konzentration 

Von den vier Variablen (A, $\alpha$, l, c) liefert das Spektrometer A die Absorbtion sowie l die Weglänge, wobei l jeweils bekannt ist bzw. durch das Gerät eingestellt wird und keine "gemessene" Grösse ist. 

Um ein Gefühl für lineare Abhängigkeiten, deren graphische Darstellung und eine erste kleine Einführung in die Programmiersprache Python zu erhalten dienen folgende "Zellen" bzw. Abschnitte in diesem Notebook.


```python
# Kommentar 1: Definiere die Variable "alpha" als eine Zahl (in diesem Fall einen sogenannten Float bzw. Gleitkommazahl)
alpha = 1.000

# Kommentar 2: Die Variable "A" oder in unserem Beispiel die Absorbtion wird als eine Aneinanderreihung von mehreren Messpunktion definiert.
# Python liefert für diesen Zweck die Möglichkeit einen sogennanten Array zu definierten.

A = [0,1,2,3,4,5]

# Kommentar 3: Die Variable "l" die Weglänge wird ebenfalls als Array definiert. Folgend wird genau der gleich Array wie in der Variablen A 
# definiert. Benutzt wird die range funktion. Mit dem Startwert 0, Stopwert 6 und Schrittwert 1. Ergebnis dieser Funktion ist der Array
# [0,1,2,3,4,5]

l = range(0,6,1)

# Kommentar 4: Somit sind die bekannten und gemessenen Variablen A, l vom Beer-Lambert Gesetz definiert. Als nächstes werden diese graphisch
# dargestellt bzw. geplottet. Die Messpunkte A auf der Y-Achse und die zugehörige Weglänge l auf der X-Achse.

# Import einer Bibliothek für Grafiken 
import matplotlib.pyplot as plt

# Plot der Datenpunkte
plt.scatter(x=l,y=A)

# Beschriftungen
plt.ylabel('y-Achse: Absorbtion')
plt.xlabel('x-Achse: Weglänge')
plt.title('Beer-Lambert Gesetz')
```
